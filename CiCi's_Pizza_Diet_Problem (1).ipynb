# Constraint & Objective Function
## Adult Daily Minimum Nutritional Requirements Per Person

$Calories >= 2000(kcal)$

**Range for Protein**:50 - 175 (gms)

$ Protein >= 50(mg) $

**Range for Total Carb**:225-325 (gms)

$Total Carbs  >= 225 (gms)$

$Sodium <= 2300 (mg)$

**Range for Sugar**:30-50 (gms)

$Sugar >= 30$

### $objective Function$

*Weight Watchers = ww*

*Quantity of each item = volume*

$(Z) = volume[i] * ww[i] (i = 0, 1, 2, ....., n)$

## Model Building

For this part, we want to find the optimal solution at the beginning, which might give high volume for just 1 item, and this kind of optimal solution might not be possible in real life as a person can't eat only 1 item in abundance. There should be variety of items present in one's meal to get full benefit from the food. Hence, we would also be building a second model that will incorporate other items.

### Model 1
 
For the first model we build, we only want to get the minimum weight watchers by meeting the requirements of daily nutritional limitation of a grown adult.But if we only use the minimum nutritional requirements as our constraints to build the model, the optimal value we get from the model solver doesn’t make sense in the real world.

We will need to imrpove our model by adding constraints or set up limitations of decision variables.

```
# declare the model
model = ConcreteModel()

# declare decision variables
model.volume = Var(products,domain=NonNegativeReals)

# declare objective

obj_expr = 0
for product in products:
  obj_expr += weightwatchers[product]*model.volume[product]
model.weight = Objective(
                      expr = obj_expr,
                      sense = minimize)

# Declare constraints

calories_expr = 0
for product in products:
  calories_expr += calories[product]*model.volume[product] 
model.Calories = Constraint(expr = calories_expr >= 2000)

protein_expr = 0
for product in products:
  protein_expr += Protein[product]*model.volume[product] 
model.protein = Constraint(expr = protein_expr >= 50)

carbs_expr = 0
for product in products:
  carbs_expr += carbs[product]*model.volume[product] 
model.carbs = Constraint(expr = carbs_expr >= 230)

sodium_expr = 0
for product in products:
  sodium_expr += sodium[product]*model.volume[product] 
model.sodium = Constraint(expr = sodium_expr <= 2300)

sugar_expr = 0
for product in products:
  sugar_expr += sugar[product]*model.volume[product] 
model.sugar = Constraint(expr = sugar_expr >= 40)

# Show the model
model.pprint()

```

```
# solve it
SolverFactory(SOLVER, executable=EXECUTABLE).solve(model).write()
# show the results
print("weight watchers = ", model.weight())

for product in products:
  print(product_name[product],model.volume[product]())
```

### Model 2

As the Model 1 does not give optimal values for each of the food items and because it is not a complete meal and not a healthy choice, we need to improve the model. We have added constraints in our Model 2 that meet the diet requirements and one cannot have the same food all day and it is not feasible. The changes we made to the Model 2 are 

* $Total the pizza Volume <=9$

* $Total Side volume <=6$

* $Volume[i] = [0,2](i=0,1,2...n)$; $Volume.domain = Intergers$


```
# declare the model
model2 = ConcreteModel()
# declare decision variables---change the domain to positive integers
model2.volume = Var(products,domain=Integers, bounds = (0, 2))


# declare objective
obj_expr = 0
for product in products:
  obj_expr += weightwatchers[product]*model2.volume[product]
model2.weight = Objective(
                      expr = obj_expr,
                      sense = minimize)
# Declare constraints
calories_expr = 0
for product in products:
  calories_expr += calories[product]*model2.volume[product] 
model2.Calories = Constraint(expr = calories_expr >= 2000)
protein_expr = 0
for product in products:
  protein_expr += Protein[product]*model2.volume[product] 
model2.protein = Constraint(expr = protein_expr >= 50)
carbs_expr = 0
for product in products:
  carbs_expr += carbs[product]*model2.volume[product] 
model2.carbs = Constraint(expr = carbs_expr >= 230)
sodium_expr = 0
for product in products:
  sodium_expr += sodium[product]*model2.volume[product] 
model2.sodium = Constraint(expr = sodium_expr <= 2300)
sugar_expr = 0
for product in products:
  sugar_expr += sugar[product]*model2.volume[product] 
model2.sugar = Constraint(expr = sugar_expr >= 40)

# use these two constraints to balance the pizza and side
pizza_expr = 0
for product in products:
  pizza_expr += pizza[product]*model2.volume[product] 
model2.pizza = Constraint(expr = pizza_expr <= 9)

side_expr = 0
for product in products:
  side_expr += side[product]*model2.volume[product] 
model2.side = Constraint(expr = side_expr <= 6)


# Show the model

model2.pprint()
```

```
print(model2.weight())
```

```
# solve it
SolverFactory(SOLVER, executable=EXECUTABLE).solve(model2).write()
# show the results

print("---weight watchers--- = ", model2.weight())

for product in products:
  print(product_name[product],model2.volume[product]())
```
